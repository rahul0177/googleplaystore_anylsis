# Google Play Store Apps analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
from matplotlib import style
```

```python
#read the csv file
store_data=pd.read_csv("googleplaystore.csv")
data=store_data.head()
print(data)
```

```python
ctg=store_data.Category.value_counts()
print(ctg)
```

```python
# plotting bar graph between Category of apps and their counts to analyze which category has maximum count
store_data.Category.value_counts().plot(kind='bar',figsize= (12,8),color=['blue','orange','green','yellow','red'])
pt.xlabel("apps on google play store",fontsize=13)
pt.ylabel("Counts of their category",fontsize="13")
pt.title(" Maximum number apps belongs to Family and Games Category",fontsize="17")
```

```python
store_data.Rating.describe()
```

```python
#The most costly App in the Store
store_data[store_data.Price == store_data.Price.max()]
```

```python
store_data['Android Ver'].value_counts()
```

```python
store_data['Android Ver'].value_counts().plot(kind='barh',figsize=(13,9))
pt.xlabel("Android versions apps counts")
pt.ylabel("Versions")
pt.title("Count Plot of the various Versions")
```
